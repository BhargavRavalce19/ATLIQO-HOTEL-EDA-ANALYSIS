```python
df_agg_bookings['occ_pct'] = (df_agg_bookings.successful_bookings / df_agg_bookings.capacity) * 100
df_agg_bookings.head()
```

**AD HOC Analysis (AD HOC Question from Manager)**
1. What is an average occupancy rate in each of the room categories ?

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean().round(2)
```

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on = "room_category", right_on = "room_id")
df.head()
```

```python
df.groupby("room_class")["occ_pct"].mean().round(2)
```

```python
df.drop("room_id", axis = 1, inplace = True)
df.head()
```

2.Print average occupancy rate per city

```python
df = pd.merge(df, df_hotels, on = "property_id")
df.head()
```

```python
df.groupby("city")["occ_pct"].mean().round(2)
```

```python
df.groupby("city")["occ_pct"].mean().round(2).plot(kind = "barh")
```

3. When was the occupamcy better ? Weekday or Weekend ?

```python
df = pd.merge(df, df_date, left_on="check_in_date", right_on="date")
df.head()
```

```python
df.groupby("day_type")["occ_pct"].mean().round(2).plot(kind = "barh")
```

4. In the month of June, what is the occupancy of different cities  

```python
df.groupby("mmm yy")["occ_pct"].mean().round(2)
```

```python
df_june_22 = df[df["mmm yy"] == "Jun 22"]
```

```python
df_june_22.occ_pct.mean().round(2)
```

```python
df_june_22.groupby("city_y")["occ_pct"].mean().round(2)
```
